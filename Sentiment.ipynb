```python
import tweepy
from textblob import TextBlob
import matplotlib.pyplot as plt

#calculating the percentage
def percentage(num,tot):
    return 100* float(num)/float(tot)


#Keys(You need to create a Twitter develpoer account inorder to get this and paste in your own keys here)
consumerkey = ""
consumersecret = ""
accesstoken = ""
accesstokensecret = ""


#establishing connection
auth = tweepy.OAuthHandler(consumerkey,consumersecret)
auth.set_access_token(accesstoken,accesstokensecret)
api = tweepy.API(auth)


#inputting the search term
print("Enter the search term or a hastag to find:")
searchterm = input()
print("\n")
print("Enter the number of tweets to be searched:")
count = int(input())
print("\n")

#finding tweets
tweets = tweepy.Cursor(api.search, q=searchterm, lang="en", since = "2020-05-10", until = "2020-10-23" ).items(count)

pos = 0
neg = 0
neut = 0
polarity = 0

for tweet in tweets:
    #print(tweet.text)
    analysis = TextBlob(tweet.text)
    polarity += analysis.sentiment.polarity
    
    
    if (analysis.sentiment.polarity == 0):
        neut += 1
    
    elif (analysis.sentiment.polarity < 0.00):
        neg += 1    

    elif (analysis.sentiment.polarity > 0.00):
        pos += 1
    
print(polarity)    
#calculating the percentages    
pos = percentage(pos,count)
#print(pos)
#print("\n")
neg = percentage(neg,count)
#print(neg)
#print("\n")
neut = percentage(neut,count)
#print(neut)
#print("\n")
polarity = polarity/count
#print(polarity)
#print("\n")

#Reducing to 2 decimal places
pos = format(pos,'.2f')
neg = format(neg,'.2f')
neut = format(neut,'.2f')


#printing the peoples opinion
print("Peoples opinion on UFC254 on " + searchterm + " winning the fight after analyzing " + str(count) +" tweets is:")
print("\n")
if(polarity == 0):
    print("Tough fight, Cannot predict the winner")
elif(polarity < 0.00):
    print("People think "+ searchterm + " will Lose")
elif(polarity > 0.00):
    print("People think "+ searchterm + " will smash")
    
#Plotting a pie chart
labels = ['Positive['+str(pos)+'%]','Neutral['+str(neut)+'%]','Negative['+str(neg)+'%]']
sizes = [pos,neut,neg]
colors = ['green' , 'orange' , 'red']
patches, texts = plt.pie(sizes, colors= colors, startangle = 90)
plt.legend(patches, labels, loc="best")
plt.title("UFC 254")
plt.axis('equal')
plt.tight_layout()
plt.show
```
